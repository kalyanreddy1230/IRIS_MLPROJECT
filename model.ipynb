```python
import pycaret
```

```python
# loading sample dataset from pycaret dataset module
from pycaret.datasets import get_data
data = get_data('diabetes')
```

```python
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
data['Class variable'].value_counts()
```

```python
# import pycaret classification and init setup
from pycaret.classification import *
s = setup(data, target = 'Class variable', session_id = 123, use_gpu=True)
```

```python
# import ClassificationExperiment and init the class
from pycaret.classification import ClassificationExperiment
exp = ClassificationExperiment()
```

```python
# check the type of exp
type(exp)
```

```python
# init setup on exp
exp.setup(data, target = 'Class variable', session_id = 123, use_gpu=True)
```

```python
# compare baseline models
best = compare_models()
```

```python
# compare models using OOP
exp.compare_models()
```

```python
# plot confusion matrix
plot_model(best, plot = 'confusion_matrix')
```

```python
# plot AUC
plot_model(best, plot = 'auc')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
# check docstring to see available plots 
# help(plot_model)
```

```python
evaluate_model(best)
```

```python
# predict on test set
holdout_pred = predict_model(best)
```

```python
# show predictions df
holdout_pred.head()
```

```python
# copy data and drop Class variable

new_data = data.copy()
new_data.drop('Class variable', axis=1, inplace=True)
new_data.head()
```

```python
# predict model on new_data
predictions = predict_model(best, data = new_data)
predictions.head()
```

```python
# save pipeline
save_model(best, 'my_first_pipeline')
```
